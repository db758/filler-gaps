```python
total = ['adjunct', 'wh', 'subject', 'object', 'leftbranch', 'coordstruct', 'complexnp', 'bridge']

acceptable = ['object', 'bridge', 'coordstruct'] #TODO: filter specifically by acceptability values?

Dlinking = ['wh', 'leftbranch', 'object']

subjacency = ['wh', 'subject', 'complexnp', 'leftbranch']

Lmarking = ['adjunct', 'subject', 'object']
Lmarking_neg = ['adjunct', 'subject']
Lmarking_pos = ['object']

empty = ['subject', 'adjunct', 'object', 'coordstruct']
empty_neg = ['subject', 'adjunct']
empty_pos = ['object', 'coordstruct']
```

```python
df['D-link effect'] = 0

def f(row):
    if row['adapt'] in Dlinking and row['test'] in Dlinking and row['adapt']!=row['test']:
        return row['corrected_diff']

df['D-link effect'] = df.apply(f, axis=1)
```

```python
df_comp['D-link comp effect'] = 0

comp = (set(total)).difference(set(Dlinking))

def f(row):
    if row['adapt'] in comp and row['test'] in Dlinking and row['adapt']!=row['test']:
        return row['corrected_diff']

df_comp['D-link comp effect'] = df_comp.apply(f, axis=1)
```

```python
df['subjacency effect'] = 0

def f(row):
    if row['adapt'] in subjacency and row['test'] in subjacency and row['adapt']!=row['test']:
        return row['corrected_diff']

df['subjacency effect'] = df.apply(f, axis=1)
```

```python
df_comp['subjacency comp effect'] = 0

comp = (set(total)).difference(set(subjacency))

def f(row):
    if row['adapt'] in comp and row['test'] in subjacency and row['adapt']!=row['test']:
        return row['corrected_diff']

df_comp['subjacency comp effect'] = df_comp.apply(f, axis=1)
```

```python
df['L-marking effect'] = 0

def f(row):
    if row['adapt'] in Lmarking:
        if row['adapt'] in Lmarking_neg and row['test'] in Lmarking_neg and row['adapt']!=row['test']:
            return row['corrected_diff']
        elif row['adapt'] in Lmarking_pos and row['test'] in Lmarking_pos and row['adapt']!=row['test']:
            return row['corrected_diff']
        
df['L-marking effect'] = df.apply(f, axis=1)
```

```python
df_comp['L-marking comp effect'] = 0

comp = (set(total)).difference(set(Lmarking))

def f(row):
    if row['adapt'] in comp and row['test'] in Lmarking:
        return row['corrected_diff']

df_comp['L-marking comp effect'] = df_comp.apply(f, axis=1)
```

```python
df['L-marking effect blue'] = 0

def f(row):
    if row['adapt'] in Lmarking:
        if row['adapt'] in Lmarking_neg and row['test'] in Lmarking_pos:
            return row['corrected_diff']
        elif row['adapt'] in Lmarking_pos and row['test'] in Lmarking_neg:
            return row['corrected_diff']

df['L-marking effect blue'] = df.apply(f, axis=1)
```

```python
df_comp['L-marking comp effect blue'] = 0

comp = (set(total)).difference(set(Lmarking))

def f(row):
    if row['adapt'] in comp and row['test'] in Lmarking:
        return row['corrected_diff']

df_comp['L-marking comp effect blue'] = df_comp.apply(f, axis=1)
```

```python
df['ECP effect'] = 0

def f(row):
    if row['adapt'] in empty:
        if row['adapt'] in empty_neg and row['test'] in empty_neg and row['adapt']!=row['test']:
            return row['corrected_diff']
        elif row['adapt'] in empty_pos and row['test'] in empty_pos and row['adapt']!=row['test']:
            return row['corrected_diff']

df['ECP effect'] = df.apply(f, axis=1)
```

```python
df_comp['ECP comp effect'] = 0

comp = (set(total)).difference(set(empty))

def f(row):
    if row['adapt'] in comp and row['test'] in empty:
        return row['corrected_diff']

df_comp['ECP comp effect'] = df_comp.apply(f, axis=1)
```

```python
df['ECP effect blue'] = 0

def f(row):
    if row['adapt'] in empty:
        if row['adapt'] in empty_neg and row['test'] in empty_pos:
            return row['corrected_diff']
        elif row['adapt'] in empty_pos and row['test'] in empty_neg:
            return row['corrected_diff']

df['ECP effect blue'] = df.apply(f, axis=1)
```

```python
df_comp['ECP comp effect blue'] = 0

comp = (set(total)).difference(set(empty))

def f(row):
    if row['adapt'] in comp and row['test'] in empty:
        return row['corrected_diff']

df_comp['ECP comp effect blue'] = df_comp.apply(f, axis=1)
```
